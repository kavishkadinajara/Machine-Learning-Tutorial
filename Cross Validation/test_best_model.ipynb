## TEST BEST MODEL TRAIN WITH RANDOM DATA

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Iris.csv")
data = data.drop("Id", axis=1)
data.head()
```

```python
x = data.drop("Species", axis=1)
x.head()
```

```python
y = data["Species"]
y.head()
```

```python
from sklearn.model_selection import train_test_split # type: ignore
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier()
modelDT.fit(x_train, y_train)
modelDT.score(x_test, y_test)
```

## K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
modelKN = KNeighborsClassifier(n_neighbors=1)
modelKN.fit(x_train, y_train)
modelKN.score(x_test, y_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
modelLR = LogisticRegression()
modelLR.fit(x_train, y_train)
modelLR.score(x_test, y_test)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
modelNB = GaussianNB()
modelNB.fit(x_train, y_train)
modelNB.score(x_test, y_test)
```

 ## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier()
modelRF.fit(x_train, y_train)
modelRF.score(x_train, y_train)
```

## Support Vector

```python
from sklearn.svm import SVC
modelSV  = SVC(kernel='linear', C=40)
modelSV.fit(x_train, y_train)
modelSV.score(x_test, y_test)
```
